```python
# Handling Missing values for reviews_per_month by replacing it with mode
def review_scan(x):
    if isinstance(x,int):
        return x
    else:
        return df.reviews_per_month.mode()
```

```python
df['reviews_per_month']=df['reviews_per_month'].apply(review_scan)
```

```python
# Handling Missing values for Last_review by dropping the column from the table.
#df.drop(['last_review'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```
